```python
import matplotlib.pyplot as plt
from glob import glob
import pandas as pd
from IPython.display import display
from auto_tqdm import tqdm
import numpy as np
```

```python
path = "dataset/original/run"
```

```python
paths = list(glob("{path}/**/history.json".format(path=path), recursive=True))
histories = pd.DataFrame([], columns=pd.read_json(paths[0]).columns)
for history in tqdm(paths):
    df = pd.read_json(history)
    df.index = df.index.astype(int)
    df = df.sort_index()
    values = df.iloc[df.index[-1]]
    values.name = int(history.split("/")[-4])
    histories = histories.append(values)
histories
histories.index = histories.index.astype(int)
histories = histories.sort_index()
```

```python
for col in histories.columns:
    y = histories[col]
    x = np.log(y.index)
    print(y.max(), y.idxmax())
    plt.plot(x, y)
    plt.title(col)
    plt.show()
```
